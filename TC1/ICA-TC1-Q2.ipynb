<h3><b>
2.1 Formulação do modelo estatístico e estimação de parâmetros
</b></h3>

O modelo adotado foi o linear múltiplo, análogo ao estudado na questão 01, mas em vez de:

$$y=ax+b$$

Tem-se:

$$y=\sum_{i=1}^{p} a_ix_i +b$$

Onde $p$ é o número de parâmetros, $x_i$ é a $i$-ésima variável independente e $b$ o intercepto. Para escrever numa forma vetorial, ou mais compacta, fazemos:

$$x_0=1$$
$$a_0=b$$

$$\vec{x}=[x_0, x_1,\dots,x_p]^t$$
$$\vec{a}=[a_0, a_1,\dots,a_p]^t$$

$$y=\vec{a}^{\hspace{.05cm}t}\vec{x}$$

Considerando um erro aleatório aditivo gaussiano, obtem-se:

$$y=\vec{a}^{\hspace{.05cm}t}\vec{x}+\epsilon$$

Onde $\epsilon$ representa essa variável aleatória.

Para a estimação dos parâmetros, ou seja, estimação do vetor $\vec{a}$, considera-se o número de amostras igual a $n$ e constroem-se as variáveis:

$$\vec{y}_{nx1}=[y_1,\dots,y_n]^t$$

$$
\tilde{X}_{nx(p+1)} =
\begin{bmatrix}
    1 & x_{11} & x_{12} & \dots  & x_{1p} \\
    1 & x_{21} & x_{22} & \dots  & x_{2p} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_{n1} & x_{n2} & \dots  & x_{np}
\end{bmatrix}
$$

Onde $y_i$ é o valor de saída da $i$-ésima amostra e $x_{ij}$ é o valor da $j$-ésima variável de entrada da $i$-ésima amostra. Dessa forma reescrevemos a equação original:

$$\vec{y}_{nx1}=\tilde{X}_{nx(p+1)}\vec{a}_{(p+1)x1}+\vec{\epsilon}_{nx1}$$

Onde $\vec{\epsilon}$ é uma variável aletória de dimensão $n$. De forma a minimizar a norma euclidiana do erro temos como solução `\cite{boyd}`:

$$\vec{a}=\tilde{X}^{\dagger} \vec{y}$$

Onde $\tilde{X}^{\dagger}$ representa a pseudo-inversa da matriz $\tilde{X}$.

Como as variáveis possuem escalas diferentes, que podem ser vistas na `\ref{tab:statsDataset}`, faz-se necessário o uso de normalização. A metodologia utilizada foi `\cite{https://en.wikipedia.org/wiki/Normalization_(statistics)}`:

$$u_i=\frac{x_i-\hat{\mu}_i}{\hat{\sigma}_i}$$

Onde $u_i$ é o novo valor da variável $x_i$, $\hat{\mu}_i$ é a estimativa da média e $\hat{\sigma}_i$ a estimativa do desvio-padrão da $i$-ésima variável de entrada.

O código abaixo aplica a normalização das variáveis e a regressão linear múltipla:

```python
# Normalização das variáveis de entrada
x_normal = (x_data - x_data.mean()) / x_data.std()
tab2 = x_normal.describe()
tab2 = tab2.loc[["mean", "std", "min", "max", "25%", "50%", "75%"],:]
tab2
```

```python
# Ordinary Least Squares
from numpy.linalg import pinv

def ols(x, y):
    n = len(y) # tamanho do dataset

    # y = Xa
    ones = np.ones((n, 1))
    X = np.concatenate((ones, x), axis=1)

    # pseudo-inversa de X
    X_cross = pinv(X)

    # a = X_cross*y
    a = np.matmul(X_cross, y)
    return a

a = ols(x_normal.values, y_data.values)
print("O parâmetros ótimos do modelo linear são: \na = {}".format(np.squeeze(a)))
```

```python
# Comparação com scikit-learn
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit (x_normal, y_data)
temp = np.concatenate((reg.intercept_, reg.coef_[0]))
print("a_sklearn = {}".format(temp))
```

<h3><b>
2.2 Análise dos resíduos
</b></h3>

Devido à dificuldade em analisar visualmente um modelo com 13 variáveis de entrada, a metodologia adotada para validar o modelo foi:
  * Cálculo do Coeficiente de Determinação ($R^2$);
  * Análise do Histograma dos resíduos com teste de hipótese de Shapiro-Wilk para normalidade;
  * Análise do Boxplot dos resíduos para identificação de possíveis outliers;
  * Análise de sensibilidade do modelo aos outliers.
  
O código abaixo foi utlizado para gerar o histograma e boxplot dos resíduos assim como calcular o valor de $R^2$ e aplicar o teste de Shapiro-Wilk:

```python
import plotly.offline as plt
import plotly.graph_objs as go
from scipy import stats

plt.init_notebook_mode(connected=True) # habilitando plot dentro do jupyter notebook

# Função que calcula o R2
def r2(y, e_vec):
    y_mean = np.mean(y)
        
    SQ_e = np.dot(np.transpose(e_vec), e_vec)          # sum of squared errors
    S_yy = sum([(y_i - y_mean)**2 for y_i in y]) # sum of errors for naive predictor

    R2 = 1 - SQ_e/S_yy
    return R2

def plotHistogram(e_vec):
    # Histograma dos resíduos
    data = [go.Histogram(
        x=e_vec,
        histnorm="probability density",
    )]

    titlefont=dict(family='Courier New, monospace', size=18, color='#7f7f7f')
    layoutHist = go.Layout(
        title='Histograma dos resíduos',
        xaxis=dict(title='Valor do erro', titlefont=titlefont),
        yaxis=dict(title='Densidade de probabilidade', titlefont=titlefont)
    )

    fig = go.Figure(data=data, layout=layoutHist)
    plt.iplot(fig)

# Boxplot dos resíduos
def plotBoxplot(e_vec):
    data = [
        go.Box(
            y= e_vec,
            boxpoints='all',
            boxmean='sd',
            jitter=0.3,
            pointpos=-1.8
        )
    ]
    
    layoutBoxplot = go.Layout(title="Boxplot dos resíduos")

    fig = go.Figure(data=data, layout=layoutBoxplot)
    plt.iplot(fig)

# criando a matriz X_normal em função do dataframe x_normal
n = len(y_data)
ones = np.ones((n, 1))
X_normal = np.concatenate((ones, x_normal.values), axis=1)

y_pred = np.dot(X_normal, a)
e_vec = np.subtract(y_data.values, y_pred)
e_vec = e_vec.transpose()[0]



plotHistogram(e_vec)
plotBoxplot(e_vec)
R2 = r2(y_data.values, e_vec)
print("R2: {}".format(R2[0]))
print("P-value: {}".format(stats.shapiro(e_vec)[1]))
```

Duas suposições do modelo se concretizaram:
  * Os resíduos apresentaram média 0 (considerando $10^{-15}$ como praticamente zero);
  * A hipótese de normalidade foi aceita após o teste de Shapiro-Wilk.

A hipótese da presença de outliers foi rejeitada pois no boxplot não foi encontrado uma ou mais amostras com erro discrepante quando comparada com as outras.

O valor de $R^2$ foi relativamente alto, mostrando que o modelo linear teve um bom ajuste nos dados.

De forma simplificada podemos analisar a importância de cada variável preditora pelo módulo do seu coeficiente multiplicativo, visto que as variáveis de entrada foram normalizadas. Para tal o código abaixo gerou a Tabela \ref{tab:modCoef}:

Agora, se quisermos analisar que variáveis são mais importantes para a construção do modelo devemos estabelecer uma metodologia de análise. Para isso foi calculado o $R^2$ e feito o teste de Shapiro-Wilk para cada combinação possível das variáveis de entrada. O código abaixo aplica o teste extensivo e apresenta os resultados na Tabela \ref{tab:analiseDeSensibilidade}:

```python
import itertools

n = x_data.shape[1] # inicialmente 13 variáveis de entrada
opcoes = list(map(list, itertools.product([0, 1], repeat=n))) # todas as combinações possíveis das variáveis

print("Número de possibilidades: {}".format(len(opcoes)))
```

```python
data = []
for opcao in opcoes:
    # retirar variáveis que serão utilizadas
    indices = [i for i, x in enumerate(opcao) if x == 1] # salvando índices em que opcao[i]=1
    x = x_normal.iloc[:,indices] # pegando as colunas referentes aos índices em que o opcao[i]=1

    a = ols(x.values, y_data.values) # parâmetros ótimos da regressão

    # calcular e_vec
    N = len(y_data) # quantidade de amostras
    X = np.concatenate((np.ones((N, 1)), x.values), axis=1)
    y_pred = np.dot(X, a)

    e_vec = np.subtract(y_data.values, y_pred).transpose()[0]

    # R2 e Shapiro-Wilk
    R2 = r2(y_data.values, e_vec)
    sw = stats.shapiro(e_vec)[1]

    # salvando informações
    data.append(opcao + [R2[0], sw])
```

```python
cabecalho = newNames + ["R2", "SW P-value"]
analiseDeSensibilidade = pd.DataFrame(data, columns=cabecalho) # dataframe com os resultados das simulações
analiseDeSensibilidade = analiseDeSensibilidade.sort_values("R2", ascending=False)
analiseDeSensibilidade.head()
```

Os cinco melhores resultados, "rankeados" pelo valor de $R^2$, estão contidos na Tabela \ref{tab:analiseDeSensibilidade}. Nessa tabela, na coluna referente a cada variável, um '1' representa a presença da dita variável no modelo e um '0' sua ausência. Conclui-se que para obter o melhor valor de $R^2$ precisa-se de todas variáveis de entrada, no entanto, podemos estudar o *trade-off* que existe entre o número de variáveis e o valor de $R^2$. Para isso formularemos o problema de otimização:

$$\text{minimizar}\hspace{0.5cm} f_o(x)=\big(-R^2(x), x\big)$$
$$\text{sujeito a:} \hspace{0.5cm} 0 \leq x \leq 1 $$

Onde $x$ é o número de variáveis normalizado:

$$x=\frac{\text{número de variáveis utilizadas}}{\text{número total de variáveis}}$$

Perceba que o valor negativo em $R^2$ indica que pretendemos maximizar essa métrica, ou seja, minimizar seu negativo assim como  minimizar $x$, o número de variáveis normalizado.

Foi feito o gráfico da região de possíveis valores de $f_o(x)$, que pode ser visto da Figura \ref{fig:grafOtimo}:

```python
import copy
data_temp = copy.deepcopy(data)

R2_points = []
nVar_points = []

for linha in data_temp:
    R2_points.append(-1*linha[-2])
    nVar_points.append(sum(linha[:-2])/len(linha[:-2]))
```

```python
traceData = go.Scatter(
    x = R2_points, 
    y = nVar_points, 
    mode='markers',
    name='dataset')
layoutData = go.Layout(
    title = "Gráfico da otimização",
    xaxis=dict(title="-R2"),
    yaxis=dict(title="Número de variáveis normalizado"))

fig = go.Figure(data=[traceData], layout=layoutData)
plt.iplot(fig)
```

Na Figura \ref{fig:grafOtimo} não existe um ponto que possua simultâneamente o menor valor de $x$ e de $-R^2$, o que existe é uma fronteira de Pareto \cite{boyd}. Para encontrar um único ponto ótimo será necessário escalonar o problema, assumindo o formato:

$$\text{minimizar}\hspace{0.5cm} f_{o_2}(x)=\lambda^t\big(-R^2(x), x\big)=-\lambda_1R^2+\lambda_2x$$
$$\text{sujeito a:} \hspace{0.5cm} 0 \leq x \leq 1 $$

Como $x$ e $R^2$ estão ambos no intervalo $[0,1]$ é coerente estabelecer um $\gamma \in [0,1]$ para representar o *trade-off* entre os objetivos, obtendo-se:

$$\text{minimizar}\hspace{0.5cm} f_{o_2}(x)=-(1-\gamma)R^2+\gamma x$$
$$\text{sujeito a:} \hspace{0.5cm} 0 \leq x \leq 1 $$

Assim para $\gamma=1$ o problema de otimização se torna minimizar apenas o número de variáveis de entrada, resultando num modelo linear contendo apenas o intercepto. Quando $\gamma=0$ o problema se torna maximizar o valor de $R^2$ que, como visto anteriormente, tem como resposta usar todas as variáveis possíveis de entrada.

Como $x$ só pode assumir 8192 valores discretos, podemos avaliar a função $f_{o_2}(x)$ para cada ponto do espaço realizável. Como exemplo adotamos $\gamma=0,5$ e o resultado pode ser visto na Tabela \ref{tab:tabGamma5}.

```python
import copy
import ipywidgets as widgets
from IPython.display import clear_output
from IPython.display import display, HTML

# função objetivo
def fo(R2, nVar, gamma):
    return -R2*(1-gamma) + gamma*nVar # gamma atribui peso, em porcentagem ao número de variáveis

# Widgets que permitem mudar o valor de gamma
gammaTxt = widgets.FloatText(
    value=0.5,
    min=0,
    max=1,
    step=0.05,
    description='gamma:'
)
gammaSlide = widgets.FloatSlider(
    value=0.5,
    min=0,
    max=1,
    step=0.05
)

display(gammaTxt, gammaSlide)
mylink = widgets.jslink((gammaTxt, 'value'), (gammaSlide, 'value'))

cabecalho = newNames + ["R2", "SW P-value", "fo"]

def acharOtimo(change):
    clear_output()
    display(gammaTxt, gammaSlide)
    gamma = change['new'] # novo número de amostras
    
    data_temp = copy.deepcopy(data)

    for linha in data_temp:
        R2 = linha[-2] # de 0 a 1
        nVar = sum(linha[:-2])/len(linha[:-2]) # número de variáveis normalizado
                                               # de 0 a 1
        linha.append(fo(R2, nVar, gamma))

    analiseDeSensibilidade = pd.DataFrame(data_temp, columns=cabecalho) # dataframe com os resultados das simulações
    analiseDeSensibilidade = analiseDeSensibilidade.sort_values("fo", ascending=True)
    display(analiseDeSensibilidade.iloc[0:5])

gammaTxt.observe(acharOtimo, names='value')
gammaSlide.observe(acharOtimo, names='value')

acharOtimo({'new': 0.5})
```

Para valores diferentes de $\gamma$ podemos achar diferentes ótimos, como exemplos tem-se:
  * $\gamma > 0,85$: a solução ótima é não utilizar nenhuma variável de entrada. O modelo linear se reduz ao intercepto, onde intercepto = média da variável de saída;
  * $0,45 \leq \gamma \leq 0,85$: a solução ótima é a que usamos apenas a circunferência do abdômen como variável de entrada;
  * $0,15 \leq \gamma \leq 0,45$: a variável peso aparece junto com a circunferência do abdomen;
  * $\gamma < 0,15$: as outras variáveis de entrada vão aparecendo até que todas fiquem presentes no modelo.

O resultado foi análogo ao anterior, e podemos fazer uma comparação do $\gamma$ com as ordens de grandezas dos coeficientes do modelo completo. No modelo orginal o coeficiente da circunferência do abdômen está uma ordem de grandeza maior que o segundo colocado, o peso, assim percebemos que para uma faixa grande de valores de $\gamma$ o melhor modelo permanece o com uma única variável de entrada. Já o peso, o segundo maior coeficiente no modelo original, tem uma faixa relativamente grande de valores de $\gamma$, visto que seu coeficiente é quase duas vezes o do terceiro lugar, a circunferência do pulso.
